```python
import numpy as np
import os, os.path
import pandas as pd
import matplotlib.pyplot as plt
from collections import OrderedDict, defaultdict
```

```python
class NestedDefaultDict(defaultdict):
    def __init__(self, *args, **kwargs):
        super(NestedDefaultDict, self).__init__(NestedDefaultDict, *args, **kwargs)

    def __repr__(self):
        return repr(dict(self))
```

```python
symbols= os.listdir(LabelsAlternateTwo)
dfs = []
dfs_dict= dict()

symbolIdx = 5
for symbolIdx in range(0, 39):
    symbolPath = os.path.join(LabelsAlternateTwo, symbols[symbolIdx])




    descriptive_dict =  NestedDefaultDict()

    for fileIdx in range(len(os.listdir(symbolPath))):
        filePath = os.path.join(symbolPath, os.listdir(symbolPath)[fileIdx])
        df = pd.read_csv(filePath)
        date = filePath.split("/")[-1].split(".")[0]

        no_trades= len(df)
        duration= np.median(df.Duration)
        return_std = np.std(df.ReturnTradedPrice)
        volume = np.median(df.Volume)

        descriptive_dict[date]['Duration']= duration
        descriptive_dict[date]['NoTrades']= no_trades
        descriptive_dict[date]['ReturnStDev']= return_std*1000
        descriptive_dict[date]['TradedVolume']= volume

        df_across_all_days = pd.DataFrame(pd.DataFrame.from_dict(descriptive_dict).median(axis=1).T)
        df_across_all_days = df_across_all_days.rename(columns={0:str(symbols[symbolIdx])})
        dfs.append(df_across_all_days)

    print(symbols[symbolIdx])
    print(pd.concat(dfs, axis =1).median(axis=1)    )
    dfs_dict[symbols[symbolIdx]] = pd.concat(dfs, axis =1).median(axis=1)    
 
 
```

```python
df_all_symbols_description = pd.DataFrame.from_dict(dfs_dict).T

print(df_all_symbols_description.median(axis=0).to_latex())
```
