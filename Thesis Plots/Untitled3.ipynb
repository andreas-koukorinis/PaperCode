```python

def gaussian_2d(x, y, sigma_x, sigma_y):
    return (1 / (2 * np.pi * sigma_x * sigma_y)) * np.exp(-((x**2)/(2*sigma_x**2) + (y**2)/(2*sigma_y**2)))

def create_kernel(size, sigma_x, sigma_y):
    ax = np.linspace(-(size // 2), (size // 2), size)
    xx, yy = np.meshgrid(ax, ax)
    kernel = gaussian_2d(xx, yy, sigma_x, sigma_y)
    return kernel

size = 51
sigma_isotropic = 10

# Isotropic Kernel
isotropic_kernel = create_kernel(size, sigma_isotropic, sigma_isotropic)

# Anisotropic (Non-Isotropic) Kernel
sigma_x = 5
sigma_y = 15
anisotropic_kernel = create_kernel(size, sigma_x, sigma_y)

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.imshow(isotropic_kernel, cmap='jet')
ax1.set_title(f'Isotropic Kernel (σx = σy = {sigma_isotropic})', loc= 'center')
ax1.axis('off')

ax2.imshow(anisotropic_kernel, cmap='jet')
ax2.set_title(f'Anisotropic Kernel (σx = {sigma_x}, σy = {sigma_y})', loc= 'center')
ax2.axis('off')
title = 'isotropic_vs_anisotropic1.png'
filePath = os.path.join(folder, title)
plt.savefig(filePath, dpi = 300)  
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Isotropic: Laplacian of Gaussian (LoG) Kernel
def log_kernel(size, sigma):
    ax = np.linspace(-(size // 2), (size // 2), size)
    xx, yy = np.meshgrid(ax, ax)
    kernel = - (1 - (xx**2 + yy**2) / (2 * sigma**2)) * np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
    return kernel / kernel.sum()

size = 7
sigma = 1.4
isotropic_kernel = log_kernel(size, sigma)

# Anisotropic: Sobel Kernels
sobel_x = np.array([[-1, 0, 1], 
                    [-2, 0, 2], 
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1], 
                    [ 0,  0,  0], 
                    [ 1,  2,  1]])

# Plot with Seaborn's heatmap
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

sns.heatmap(isotropic_kernel, cmap='coolwarm', center=0, ax=ax1)
ax1.set_title(f'Isotropic: LoG Kernel (σ = {sigma})')
ax1.axis('off')

sns.heatmap(sobel_x, cmap='coolwarm', center=0, ax=ax2)
ax2.set_title('Anisotropic: Sobel X Kernel')
ax2.axis('off')

sns.heatmap(sobel_y, cmap='coolwarm', center=0, ax=ax3)
ax3.set_title('Anisotropic: Sobel Y Kernel')
ax3.axis('off')
title = 'isotropic_vs_anisotropic1.png'
filePath = os.path.join(folder, title)
plt.savefig(filePath, dpi = 300)  
plt.tight_layout()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for demonstration
num_levels = 10

best_bid_price = 10.0
best_ask_price = 10.1  # this can be set to whatever value is representative of the market spread

bid_prices = [best_bid_price - (i**2)*0.01 for i in range(num_levels)]
ask_prices = [best_ask_price + (i**2)*0.01 for i in range(num_levels)]

bid_volumes = np.random.randint(100, 500, num_levels)
ask_volumes = np.random.randint(100, 500, num_levels)

relative_bid_prices = [price - best_bid_price for price in bid_prices]
relative_ask_prices = [price - best_ask_price for price in ask_prices]

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bid and Ask Volumes
ax.bar(bid_prices, bid_volumes, width=0.02, color='g', align='center', label='$q^{\text{bid}}_{l}(t_{i})$ Volumes')
ax.bar(ask_prices, ask_volumes, width=0.02, color='r', align='center', label='$q^{\text{ask}}_{l}(t_{i})$ Volumes')

# Setting labels and titles
ax.set_ylabel('Volume')
ax.set_xlabel('Price')
ax.set_title('Limit Order Book Visualization')
ax.legend(loc='upper left')

# Secondary x-axis for relative prices
ax2 = ax.twiny()
ax2.set_xlim(ax.get_xlim())
ax2.set_xticks(bid_prices + ask_prices)
ax2.set_xticklabels(relative_bid_prices + relative_ask_prices, rotation=45, fontsize=8)
ax2.set_xlabel('Relative Price')

# Mark best bid and ask prices
ax.annotate('$p^{\text{bid}}_1(t)$', xy=(best_bid_price, 0), xycoords=('data', 'axes fraction'),
            xytext=(0, -30), textcoords='offset points', arrowprops=dict(arrowstyle='->'), ha='center')
ax.annotate('$p^{\text{ask}}_1(t)$', xy=(best_ask_price, 0), xycoords=('data', 'axes fraction'),
            xytext=(0, -30), textcoords='offset points', arrowprops=dict(arrowstyle='->'), ha='center')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for demonstration
num_levels = 10

best_bid_price = 10.0
best_ask_price = 10.1

# Generate prices in an asymmetric step-wise manner
bid_prices = [best_bid_price - (i**2)*0.01 for i in range(num_levels)]
ask_prices = [best_ask_price + (i**2)*0.01 for i in range(num_levels)]

# Generate U-shaped volumes
max_volume = 5000
min_volume = 1000
bid_volumes = [min_volume + (max_volume - min_volume) * (i/(num_levels-1)) * (1 - i/(num_levels-1)) for i in range(num_levels)]
ask_volumes = [min_volume + (max_volume - min_volume) * (i/(num_levels-1)) * (1 - i/(num_levels-1)) for i in range(num_levels)]

relative_bid_prices = [f"{price - best_bid_price:.2f}" for price in bid_prices]
relative_ask_prices = [f"{price - best_ask_price:.2f}" for price in ask_prices]

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))

# Bid and Ask Volumes
ax.bar(bid_prices, bid_volumes, width=0.02, color='g', align='center', label='$q^{\text{bid}}_{l}(t_{i})$ Volumes')
ax.bar(ask_prices, ask_volumes, width=0.02, color='r', align='center', label='$q^{\text{ask}}_{l}(t_{i})$ Volumes')

# Cumulative volumes
ax.plot(bid_prices, np.cumsum(bid_volumes), 'g--', label='Cumulative Bid Volume')
ax.plot(ask_prices, np.cumsum(ask_volumes), 'r--', label='Cumulative Ask Volume')

# Setting labels, titles, and tick formatting
ax.set_ylabel('Volume')
ax.set_xlabel('Price')
ax.set_title('Limit Order Book Visualization')
ax.legend(loc='upper left')
ax.set_xticks(bid_prices[::2] + ask_prices[::2])  # Taking every second tick to reduce overlap
ax.set_xticklabels([f"{p:.2f}" for p in bid_prices[::2]] + [f"{p:.2f}" for p in ask_prices[::2]], rotation=45, fontsize=10)

# Secondary x-axis for relative prices
ax2 = ax.twiny()
ax2.set_xlim(ax.get_xlim())
ax2.set_xticks(bid_prices[::2] + ask_prices[::2])  # Consistency with main x-axis
ax2.set_xticklabels(relative_bid_prices[::2] + relative_ask_prices[::2], rotation=45, fontsize=10)
ax2.set_xlabel('Relative Price', fontsize=12)

# Mark best bid and ask prices
ax.annotate('Best Bid', xy=(best_bid_price, min_volume), xytext=(-100, 50), textcoords='offset points', arrowprops=dict(arrowstyle='->', linewidth=1.5), ha='center', fontsize=12)
ax.annotate('Best Ask', xy=(best_ask_price, min_volume), xytext=(100, 50), textcoords='offset points', arrowprops=dict(arrowstyle='->', linewidth=1.5), ha='center', fontsize=12)

plt.tight_layout()
plt.show()
```
