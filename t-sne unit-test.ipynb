```python


from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd


# Define parameters for normal distribution.
mu_one = 5
sigma_one = 3
rng = range(-30,30)


# Define parameters for normal distribution.
mu_two = -5
sigma_two = 3
rng = range(-30,30)
# Generate normal distribution with given mean and standard deviation.
dist_one = norm(mu_one, sigma_one)
dist_two = norm(mu_two, sigma_two)

# Plot probability density function and of this distribution.
# the pdf() method takes takes in a list x values and returns a list of y's.
plt.subplot(311) # Creates a 3 row, 1 column grid of plots, and renders the following chart in slot 1.
plt.plot(rng, dist_one.pdf(rng), 'r', linewidth=2)
plt.plot(rng, dist_two.pdf(rng), 'g', linewidth=2)

plt.title('Probability density function of normal distribution')


# # Plot probability density function and of this distribution.
# plt.subplot(312)
# plt.plot(rng, dist.cdf(rng))
# plt.title('Cumulutative distribution function of normal distribution')

# Draw 1000 samples from the random variable.
sample_one = dist_one.rvs(size=10000)
sample_two = dist_two.rvs(size=10000)

print ("Sample descriptive statistics:")
print (pd.DataFrame(sample_one).describe)

# Plot a histogram of the samples.
f = plt.figure(figsize= (10,10))
ax = f.add_subplot(313)
ax.hist(sample_one, bins=100, normed=True)
ax.plot(rng, dist_one.pdf(rng), 'r--', linewidth=2)
ax.hist(sample_two, bins=100, normed=True)
ax.plot(rng, dist_two.pdf(rng), 'r--', linewidth=2)

plt.title('10,000 random samples from normal distribution')

# Show all plots.
plt.show()
```

```python
from sklearn.utils import shuffle
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler
from sklearn.pipeline import make_pipeline
import pickle as pkl
from sklearn.manifold import TSNE

from sklearn.kernel_approximation import RBFSampler
```

```python
X = np.vstack((sample_one, sample_two)).T
```

```python
scaler = MinMaxScaler(feature_range = (-1, 1))
mm = make_pipeline(scaler, Normalizer())
```

```python
rbf_sampler = RBFSampler(gamma =0.5, random_state= 10)
```

```python
for gamma in [0.5, 1.5, 2.5, 5.5, 10.5, 20.5]:
    rbf_sampler = RBFSampler(gamma =gamma, random_state= 10)
    X_transformed_one = mm.fit_transform(X)
    X_rbf_ = rbf_sampler.fit_transform(X_transformed_one)
    n_components = 2
    tsne = TSNE(n_components)
    tsne_result = tsne.fit_transform(X_rbf_)
    tsne_result_df = pd.DataFrame({'tsne_1':tsne_result[:,0],'tsne_2':tsne_result[:,1]} )
    x = tsne_result_df['tsne_1']
    y = tsne_result_df['tsne_2']

    lim = (tsne_result.min() - 5, tsne_result.max()+5)
    ax.set_xlim(lim)
    ax.set_ylim(lim)
    plt.scatter(x, y, cmap = plt.cm.rainbow)
    plt.xlabel('T-SNE 1')
    plt.ylabel('TSNE-2')
    print()
    plt.show()
```
