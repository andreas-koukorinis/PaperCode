```python
import numpy as np
from scipy.special import expit as sigmoid

def sigmoid_curve(x, x0, k):
    if x0 < 0:
        y = sigmoid((x - x0) * k)
        return y - 1
    elif x0 == 0:
        y = sigmoid((x - x0) * k)
        return 0
    else:
        return 1 - sigmoid((x - x0) * k)

# Generate symmetric sigmoid curves
x = np.linspace(-2, 2, 100)
y_ask = sigmoid_curve(x, -0.5, 5)
y_bid = sigmoid_curve(x, -0.8, 5)

# Find the intersection point
x_mid = x[np.argmin(np.abs(y_ask - y_bid))]

# Define bid and offer
bid = y_ask[np.argmin(np.abs(x - x_mid))]
offer = y_bid[np.argmin(np.abs(x - x_mid))]

# Plot curves
import matplotlib.pyplot as plt
plt.plot(x, y_ask)
plt.plot(x, y_bid)

# Add vertical lines at bid and offer
plt.axvline(x=bid, color='green', linestyle='--', label='Bid')
plt.axvline(x=offer, color='red', linestyle='--', label='Offer')

# Set labels and title
plt.xlabel('Price')
plt.ylabel('Flow')
plt.title('Symmetric Flow Curves')
plt.legend(['Ask', 'Bid', 'Bid', 'Offer'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import expit, logit

def generalized_sigmoid(x, x0, alpha, beta):

    z = (x - x0) * beta  
    y = 1 / (1 + np.exp(-z)) ** alpha
    return y

x = np.linspace(-5, 5, 100)
x0 = 0
alphas = [0.5, 1, 2]
betas = [1, 5, 10]

X, Y = np.meshgrid(x, alphas) 

fig = plt.figure()
ax = plt.axes(projection='3d')

for beta in betas:
    Z = generalized_sigmoid(X, x0, Y, beta)
    ax.plot_surface(X, Y, Z, alpha=0.4)
    
ax.set_title('Generalized Flow Curve Model')
ax.set_xlabel('X')
ax.set_ylabel('Alpha')
ax.set_zlabel('Y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random data for the empirical distribution
np.random.seed(0)
data = np.random.normal(loc=0.9, scale=1.8, size=1000)

# Fit a Gaussian distribution using MLE
mu, std = norm.fit(data)

# Create a range of values for the x-axis
x = np.linspace(-4, 4, 100)

# Calculate the fitted MLE curve
mle_curve = norm.pdf(x, mu, std)

# Create a range of values for the sigmoid curve
sigmoid_x = np.linspace(-4, 4, 100)

# Calculate the sigmoid curve (reversed) with adjusted scale
sigmoid_curve = 1 / (1 + np.exp(-10 * (sigmoid_x - mu)))

# Create a new figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot the empirical distribution as a bar plot
ax1.hist(data, bins=20, density=True, alpha=0.6, color='blue', label='Empirical')
ax1.set_title('Empirical Distribution')
ax1.set_xlabel('Charge in BP')
ax1.set_ylabel('Density')
ax1.legend()
ax1.grid(True)

# Plot the fitted MLE curve
ax1.plot(x, mle_curve, 'r-', label='Fitted MLE')
ax1.legend()

# Plot the empirical distribution bar plot
ax2.hist(data, bins=20, density=True, alpha=0.6, color='blue', label='Empirical')
ax2.set_title('Cumulative Probability ')
ax2.set_xlabel('Charge in BP')
ax2.set_ylabel('Probability to Trade')
ax2.legend()
ax2.grid(True)

# Plot the sigmoid curve with adjusted scale
ax2.plot(sigmoid_x, sigmoid_curve, 'r-', label='Sigmoid Curve')
ax2.set_ylim(0, 1)  # Set y-axis limits from 0 to 1
ax2.legend()

plt.tight_layout()
plt.savefig('/home/ak/Documents/GSR Work/FittedMLE_and_Sigmoid.png', dpi=300)
plt.show()
```
