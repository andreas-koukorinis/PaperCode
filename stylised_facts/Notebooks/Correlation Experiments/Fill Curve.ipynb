```python
import numpy as np
from scipy.special import expit as sigmoid

def sigmoid_curve(x, x0, k):
    if x0 < 0:
        y = sigmoid((x - x0) * k)
        return y - 1
    elif x0 == 0:
        y = sigmoid((x - x0) * k)
        return 0
    else:
        return 1 - sigmoid((x - x0) * k)

# Generate symmetric sigmoid curves
x = np.linspace(-2, 2, 100)
y_ask = sigmoid_curve(x, -0.5, 5)
y_bid = sigmoid_curve(x, -0.8, 5)

# Find the intersection point
x_mid = x[np.argmin(np.abs(y_ask - y_bid))]

# Define bid and offer
bid = y_ask[np.argmin(np.abs(x - x_mid))]
offer = y_bid[np.argmin(np.abs(x - x_mid))]

# Plot curves
import matplotlib.pyplot as plt
plt.plot(x, y_ask)
plt.plot(x, y_bid)

# Add vertical lines at bid and offer
plt.axvline(x=bid, color='green', linestyle='--', label='Bid')
plt.axvline(x=offer, color='red', linestyle='--', label='Offer')

# Set labels and title
plt.xlabel('Price')
plt.ylabel('Flow')
plt.title('Symmetric Flow Curves')
plt.legend(['Ask', 'Bid', 'Bid', 'Offer'])
plt.show()
```

```python
import numpy as np 
from scipy.special import expit as sigmoid

# Sigmoid function
def sigmoid_curve(x, x0, k): 
    y = sigmoid((x - x0)*k)
    return y

# Generate curves
x = np.linspace(0, 2, 100)

# Ask curve   
k_ask = 2 # Steeper slope
x0_ask = -0.5  
y_ask = sigmoid_curve(x, x0_ask, k_ask) + 0.2  

# Bid curve
k_bid = 5
x0_bid = 0.5
y_bid = sigmoid_curve(x, x0_bid, k_bid)

# Find intersection 
x_mid = x[np.argmin(np.abs(y_ask - y_bid))]
y_mid = y_ask[np.argmin(np.abs(x - x_mid))]

# Plot 
import matplotlib.pyplot as plt
plt.plot(x, y_bid, label='Bid') 
plt.plot(x, y_ask, label='Ask')

# Horizontal line at intersection
plt.hlines(y_mid, 0, 2, colors='r', linestyles='--')

plt.legend()
plt.xlabel('Price')  
plt.ylabel('Flow')
plt.title('Asymmetric Flow Curves')  

plt.show()
```

```python
import numpy as np
from scipy.special import expit as sigmoid

# Sigmoid functions 
def sigmoid_curve(x, x0, k):
    y = sigmoid((x - x0)*k) 
    return y

def reverse_sigmoid_curve(x, x0, k):
    y = 1 - sigmoid((x - x0)*k)
    return y

# Generate curves
x = np.linspace(0, 2, 100)  

# Bid curve
x0_bid = -1
k_bid = 5
y_bid = sigmoid_curve(x, x0_bid, k_bid)

# Ask curve  
x0_ask = 1  
k_ask = 5
y_ask = reverse_sigmoid_curve(x, x0_ask, k_ask)   

# Find intersection
x_mid = 1 
y_mid = 0.5

# Plot
import matplotlib.pyplot as plt
plt.plot(x, y_bid, label='Bid')
plt.plot(x, y_ask, label='Ask') 

# Vertical line at intersection 
plt.vlines(x_mid, 0, 1, colors='r', linestyles='--')

plt.legend()
plt.xlabel('Price')
plt.ylabel('Flow')
plt.title('Asymmetric Flow Curves')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit as sigmoid

# Inverse sigmoid function
def inv_sigmoid(x, x0, k):
    y = 1 - sigmoid((x - x0)*k)
    return y

# Generate curve
x = np.linspace(0, 2, 100)
x0 = 1
k = 5
y = inv_sigmoid(x, x0, k)
x0_2 = 0.8
y_2 = inv_sigmoid(x, x0_2, k)

# Plot  
plt.plot(x, y, label='Ask Flow Curve 2')
plt.plot(x, y_2, label='Ask Flow Curve 1')

# Horizontal line
plt.axhline(y=0.9, color='gray', linestyle='--', lw=2) 

plt.xlabel('Price')
plt.ylabel('Flow')
plt.title('Inverse Sigmoid Ask Flow Curve')
plt.legend()
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import expit, logit

def generalized_sigmoid(x, x0, alpha, beta):

    z = (x - x0) * beta  
    y = 1 / (1 + np.exp(-z)) ** alpha
    return y

x = np.linspace(-5, 5, 100)
x0 = 0
alphas = [0.5, 1, 2]
betas = [1, 5, 10]

X, Y = np.meshgrid(x, alphas) 

fig = plt.figure()
ax = plt.axes(projection='3d')

for beta in betas:
    Z = generalized_sigmoid(X, x0, Y, beta)
    ax.plot_surface(X, Y, Z, alpha=0.4)
    
ax.set_title('Generalized Flow Curve Model')
ax.set_xlabel('X')
ax.set_ylabel('Alpha')
ax.set_zlabel('Y')

plt.show()
```
