```python
df =testdf
def garman_klass_parkinson_vols(df, rolling_window: int = 5):
    df['log_hl'] = (df['MicroPrice_high'] / df['MicroPrice_low']).replace([np.inf, -np.inf], np.nan).apply(np.log).fillna(0)
    df['log_cc'] = (df['MicroPrice_close'] / df['MicroPrice_open']).replace([np.inf, -np.inf], np.nan).apply(np.log).fillna(0)
    df['garman_klass_vol'] = np.sqrt(0.5 * df['log_hl']**2 - (2*np.log(2) - 1) * df['log_cc']**2) 
    df['log_hl_squared'] = np.log(df['MicroPrice_high'] / df['MicroPrice_low'])**2
    df['parkinson_vol'] = np.sqrt((1 / (4 * np.log(2))) * df['log_hl_squared'].mean())
    df['bipower_var'] = (np.abs(df['log_hl']) * np.abs(df['log_hl'].shift(1))).rolling(rolling_window).sum()
    return df
```

```python
import os

def get_unique_elements(directory, subdirectory1, subdirectory2):
    path1 = os.path.join(directory, subdirectory1)
    path2 = os.path.join(directory, subdirectory2)
    
    if not os.path.isdir(path1) or not os.path.isdir(path2):
        print("Error: One or both subdirectories do not exist.")
        return []
    
    elements1 = set(os.listdir(path1))
    elements2 = set(os.listdir(path2))
    
    unique_elements = elements1.symmetric_difference(elements2)
    return unique_elements

# Usage example
directory = '/media/ak/Data/InterestRateFuturesData/QuotesTrades/KE1'
subdirectory1 = 'KE1_Comdty_quotes'
subdirectory2 = 'KE1_Comdty_trades'

unique_elements = get_unique_elements(directory, subdirectory1, subdirectory2)
print(unique_elements)
```

```python
import os

def create_subdirectories(main_directory, subdirectory_names):
    # Create the main directory if it doesn't exist
    if not os.path.exists(main_directory):
        os.makedirs(main_directory)

    # Create subdirectories within the main directory
    for name in subdirectory_names:
        subdirectory_path = os.path.join(main_directory, name)
        os.makedirs(subdirectory_path)
        print(f"Created subdirectory: {subdirectory_path}")
```
