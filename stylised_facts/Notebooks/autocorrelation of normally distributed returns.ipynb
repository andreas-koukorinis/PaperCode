```python
import numpy as np
import pandas as pd
mu, sigma = 0, 0.1 # mean and standard deviation
s = pd.np.random.normal(mu, sigma, 1000)
```

```python
from matplotlib import pyplot
```

```python
pyplot.plot(s)
pyplot.show()
```

```python
pyplot.hist(s, bins = 100)
pyplot.show()
```

```python
# create an autocorrelation plot

from statsmodels.graphics.tsaplots import plot_acf
plot_acf(s,lags=30)
pyplot.show()
```
